```python
list_year = ["Season=2021-22","Season=2020-21","Season=2019-20","Season=2018-19","Season=2017-18"]
list_gametype = ["SeasonType=Regular"]
url_base_nba_players = "https://www.basketball-reference.com/leagues/NBA_2007_games-october.html"

request_text = request.urlopen(url_base_nba_players).read()
#print(request_text[:1000])    
page = bs4.BeautifulSoup(request_text, "lxml")
print(page.body)
```

```python
tableau = page.find('table', {'class' : 'suppress_glossary sortable stats_table'})
```
