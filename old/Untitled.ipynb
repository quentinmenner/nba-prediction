```python
REQUEST_HEADERS = {
    'user-agent': 'USER_AGENT',
}
```

```python
#NBA_URL = 'http://stats.nba.com/stats/teamgamelogs'
NBA_URL = 'https://www.nba.com/stats/teams/traditional/?sort=W_PCT&dir=-1'
NBA_ID = '00'
```

```python
NBA_SEASON_TYPES = {
    'regular': 'Regular Season',
    'playoffs': 'Playoffs',
    'preseason': 'Pre Season',
}
```

```python
season = '2016-17'
season_type = NBA_SEASON_TYPES['regular']
```

```python
nba_params = {
    'LeagueID': NBA_ID,
    'Season': season,
    'SeasonType': season_type,
}
```

```python
r = requests.get(NBA_URL, params=nba_params, headers=REQUEST_HEADERS, allow_redirects=False, timeout=15)
#r = requests.get(NBA_URL, params=nba_params, allow_redirects=False, timeout=15)
r.status_code
```
