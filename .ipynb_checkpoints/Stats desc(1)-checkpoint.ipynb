```python
#donne moyenne de points marqué à l'éxtérieur (resp. domicile) chaque année (0 = 2005 etc ...), sortie d'après.
listannsdom = []
list = []

for j in range(14) :
    for i in range(16979):
        if literal_eval(df_match["Y-M"][i])[0] == 2005+j:
            list.append(df_match['PTS'][i])
    listannsdom.append(list)
    list = []
    
listannsext = []
listtt = []

for j in range(14) :
    for i in range(16979):
        if literal_eval(df_match["Y-M"][i])[0] == 2005+j:
            listtt.append(df_match['PTS.1'][i])
    listannsext.append(listtt)
    listtt = []


print("test")

means = []
for i in range(len(listannsdom)):
    mean1 = st.fmean(listannsdom[i])
    means.append(mean1)

meanss = []
for i in range(len(listannsext)):
    mean1 = st.fmean(listannsext[i])
    meanss.append(mean1)

print(means)
print(meanss)
```

```python
years = []
for i in range(14):
    years.append(2005+i)

colname = ['Moyenne de points marqués chaque année (domicile)'] 
colname2 = ['Moyenne de points marqués chaque année (extérieur)'] 

meanPTS = pd.DataFrame(means, years, colname)
meanPTS1 = pd.DataFrame(meanss, years, colname2)



display(meanPTS)
display(meanPTS1)


```

```python
#donne nb de points moyens marqués par match par les équipes 1ères de leurs conférences sur toutes les saisons
#2011 est exclu car la saison est tronquée par une grève des joueurs

meanEST = []
meanOUEST = []

for i in range(14):
        if i != 6:
            list2 = []
            list3 = []
            for k in range(16979):
                if (literal_eval(df_match["Y-M"][k])[0] == 2005+i and literal_eval(df_match["Y-M"][k])[1] >= 9) or (literal_eval(df_match["Y-M"][k])[0] == 2005+i and literal_eval(df_match["Y-M"][k])[1] <= 5):
                        if df_match['Visitor/Neutral'][k] == df_result['ouest'][i]:
                            list2.append(df_match['PTS'][k])
                        elif df_match['Home/Neutral'][k] == df_result['ouest'][i]:
                            list2.append(df_match['PTS.1'][k])
                        if df_match['Visitor/Neutral'][k] == df_result['est'][i]:
                            list3.append(df_match['PTS'][k])
                        elif df_match['Home/Neutral'][k] == df_result['est'][i]:
                            list3.append(df_match['PTS.1'][k])
            print('year',2005+i,' ',len(list2),' ',df_result['ouest'][i])
            print('year',2005+i,' ',len(list3),' ',df_result['est'][i])
            meanEST.append(st.fmean(list2))
            meanOUEST.append(st.fmean(list3))

    
print(meanEST)
print(meanOUEST)
```
