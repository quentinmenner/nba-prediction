```python
#import des répertoires dont on a besoin
import pandas as pd
from ast import literal_eval
import statistics as st
import numpy as np
from io import StringIO
```

```python
#on charge les dataframe qu'on va utiliser
df_equipes = pd.read_csv('stats-equipes-par-mois.csv', engine="python", sep=";|\r" , header= 0)
df_match = pd.read_csv('stats-matchs.csv', sep=";")
df_result = pd.read_csv('tableaupython.csv', sep=";") #correspond au dataframe stockant les premiers de chaque saison régulière de 2005 à 2019
```

```python
#cette cellulle est là car il peut y avoir des problèmes avec la fonction mean selon la version de statistics utilisée
if ('fmean' in dir(st)):
    haveFmean = True
else :
    haveFmean = False
    
def mean(x) :
    if (haveFmean) :
        return st.fmean(x)
    else :
        return st.mean(x) 
```

```python
#travail sur les stats par mois des équipes
#on essaye de déterminer chaque année le nombre de FGM (field goal par match) de l'équipe première de la saison régulière
meanOUESTTT = []
meanESTTT = []
for i in range(14):
    if i != 6:
        a = df_result['ouest'][i] #affecte à a le 1er de l'année 2005+i à l'ouest
        b = df_result['est'][i] #affecte à b le 1er de l'année 2005+i à l'est
        for k in range(1,31):
            if df_equipes.columns[k] == a:
                aN = k #affecte à aN le rang de l'équipe 1ère à l'ouest de l'année 2005+i dans les colonnes de df_equipes
            if df_equipes.columns[k] == b  :
                bN = k #affecte à bN le rang de l'équipe 1ère à l'est de l'année 2005+i dans les colonnes de df_equipes
        list1 = []
        list2 = []
        for j in range(1,98): #itère sur chaque ligne pour collecter les données de la saison 2005+ 
            if (literal_eval(df_equipes["Unnamed: 0"][j])[0] == 2005+i and literal_eval(df_equipes["Unnamed: 0"][j])[1] >= 9) or (literal_eval(df_equipes["Unnamed: 0"][j])[0] == 2006+i and literal_eval(df_equipes["Unnamed: 0"][j])[1] <= 5):#vérifie que la ligne est dans la saison que l'on souhaite étudier
                if literal_eval(df_equipes["Unnamed: 0"][j])[0] == 2006 or literal_eval(df_equipes["Unnamed: 0"][j])[0] == 2007 or literal_eval(df_equipes["Unnamed: 0"][j])[0] == 2012:
                    if literal_eval(df_equipes["Unnamed: 0"][j])[1] != 10:
                        list1.append(float(literal_eval(df_equipes[df_equipes.columns[aN]][j])['FGM']))
                        list2.append(float(literal_eval(df_equipes[df_equipes.columns[bN]][j])['FGM']))
                        #print('année',i,'valide',list1,list2,'mois',j)
                else :
                    list1.append(float(literal_eval(df_equipes[df_equipes.columns[aN]][j])['FGM']))
                    list2.append(float(literal_eval(df_equipes[df_equipes.columns[bN]][j])['FGM']))
                    #print('année',i,'valide',list1,list2,'mois',j)
                    
        meanOUESTTT.append(mean(list1))
        meanESTTT.append(mean(list2))


print(meanESTTT)
print(meanOUESTTT)
```

```python
#on calcule la somme des différences entre moyenne du premier et des non-premiers pour chaque variable

liste_écartsEST = []
liste_écartsOUEST = []
for i in range(len(liste_variables)) :
    aT = 0
    bT = 0
    for j in range(13) :
        a = liste_moyennes_1er[i][1][j] - liste_moyennes_nuls[i][1][j]
        b = liste_moyennes_1er[i][2][j] - liste_moyennes_nuls[i][2][j]
        aT = aT + a
        bT = bT + b
    liste_écartsOUEST.append([aT])
    liste_écartsEST.append([bT])
                         
colname3 = ['écart entre le 1er et le reste des variables en conférence ouest']
colname4 = ['écart entre le 1er et le rest des variables en conférence est']
                         
ECARTest = pd.DataFrame(liste_écartsEST, liste_variables, colname3)
ECARTouest = pd.DataFrame(liste_écartsOUEST, liste_variables, colname4)

display(ECARTest)
display(ECARTouest)
```

```python
#On calcule la moyenne pondérée par la moyenne totale de chaque variable (supérieure à 1 --> variable significative)
liste_écarts_pondérésOUEST = []
liste_écarts_pondérésEST = []
for i in range(len(liste_variables)) :
    liste_moyennes_pondéréesEST = []
    liste_moyennes_pondéréesOUEST = []
    for j in range(13) :
        a = (liste_moyennes_1er[i][1][j]/liste_moyennes_nuls1ER[i][1][j])
        b = (liste_moyennes_1er[i][2][j]/liste_moyennes_nuls1ER[i][2][j])
        liste_moyennes_pondéréesEST.append(a)
        liste_moyennes_pondéréesOUEST.append(b)
    liste_écarts_pondérésEST.append(mean(liste_moyennes_pondéréesEST))
    liste_écarts_pondérésOUEST.append(mean(liste_moyennes_pondéréesOUEST))

colname3 = ['moyenne pondérée des variables en conférence ouest']
colname4 = ['moyenne pondérée des variables en conférence est']
                         
PONDEREEest = pd.DataFrame(liste_écarts_pondérésEST, liste_variables, colname3)
PONDEREEouest = pd.DataFrame(liste_écarts_pondérésOUEST, liste_variables, colname4)
display(PONDEREEest)
display(PONDEREEouest)

Liste_moyenne_pondereesOUESTEST = []
for i in range(len(liste_variables)) :
    a = (float(liste_écarts_pondérésEST[i]) + float(liste_écarts_pondérésOUEST[i]))/2
    Liste_moyenne_pondereesOUESTEST.append([a])
    
colname5 = ['moyenne pondérée des variables conférences confondues']

PONDEREE = pd.DataFrame(Liste_moyenne_pondereesOUESTEST, liste_variables, colname5)
display(PONDEREE)
```
